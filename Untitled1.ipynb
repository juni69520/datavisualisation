```python
# Groupement par continent et date, et somme cumulée des morts
grouped_df = df.groupby(['continent', pd.Grouper(key='date', freq='M')])['total_cases'].sum().groupby(level=[0]).cumsum().reset_index()

# Création du graphique
fig, ax = plt.subplots(figsize=(10, 6))

# Tracé des courbes pour chaque continent
for continent, data in grouped_df.groupby('continent'):
    ax.plot(data['date'], data['total_cases'], label=continent)

# Paramètres du graphique
ax.set_title("Nombre cumulé de cas par mois et par continent")
ax.set_xlabel("Date")
ax.set_ylabel("Nombre cumulé de cas")
ax.legend()
plt.show()
```

```python
# Groupement par continent et date, et somme des nouveaux vaccinés
grouped_df = df.groupby(['continent', pd.Grouper(key='date', freq='M')])['new_vaccinations'].sum().reset_index()
grouped_df = grouped_df.query('continent != "Asia"')

# Création du graphique
fig, ax = plt.subplots(figsize=(10, 6))

# Tracé des courbes pour chaque continent
for continent, data in grouped_df.groupby('continent'):
    ax.plot(data['date'], data['new_vaccinations'], label=continent)

# Paramètres du graphique
ax.set_title("Nombre de vaccinés par mois et par continent sauf l'Asie")
ax.set_xlabel("Date")
ax.set_ylabel("Nombre de vaccinés")
ax.legend()

# Groupement par continent et date, et somme des nouveaux vaccinés
grouped_df2 = df.groupby(['continent', pd.Grouper(key='date', freq='M')])['new_vaccinations'].sum().reset_index()
grouped_df2 = grouped_df2.query('continent == "Asia"')

# Création du graphique
fig, ax2 = plt.subplots(figsize=(10, 6))

# Tracé des courbes pour chaque continent
for continent, data in grouped_df2.groupby('continent'):
    ax2.plot(data['date'], data['new_vaccinations'], label=continent)

# Paramètres du graphique
ax2.set_title("Nombre de vaccinés par mois de l'Asie")
ax2.set_xlabel("Date")
ax2.set_ylabel("Nombre de vaccinés")
ax2.legend()
plt.show()
```

```python
# Groupement par continent et date, et somme des nouvelles morts
grouped_df = df.groupby(['continent', pd.Grouper(key='date', freq='M')])['new_deaths'].sum().reset_index()

# Création du graphique
fig, ax = plt.subplots(figsize=(10, 6))

# Tracé des courbes pour chaque continent
for continent, data in grouped_df.groupby('continent'):
    ax.plot(data['date'], data['new_deaths'], label=continent)

# Paramètres du graphique
ax.set_title("Nombre de morts par mois et par continent")
ax.set_xlabel("Date")
ax.set_ylabel("Nombre de morts")
ax.legend()
plt.show()
```

```python
# Groupement par continent et date, et somme cumulée des morts
grouped_df = df.groupby(['continent', pd.Grouper(key='date', freq='M')])['total_deaths'].sum().groupby(level=[0]).cumsum().reset_index()

# Création du graphique
fig, ax = plt.subplots(figsize=(10, 6))

# Tracé des courbes pour chaque continent
for continent, data in grouped_df.groupby('continent'):
    ax.plot(data['date'], data['total_deaths'], label=continent)

# Paramètres du graphique
ax.set_title("Nombre cumulé de morts par mois et par continent")
ax.set_xlabel("Date")
ax.set_ylabel("Nombre cumulé de morts")
ax.legend()
plt.show()
```
